```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns 
df=pd.read_csv('C:\\Users\\Administrator\\Desktop\\advertising.csv')
df
```

```python
fig,(ax1,ax2,ax3)=plt.subplots(1,3, figsize=(15,6))
ax1.scatter(x=df['TV'],y=df['Sales'])
ax1.set_xlabel('TV',fontsize=12)
ax1.set_ylabel('Sales',fontsize=12)
ax2.scatter(x=df['Radio'],y=df['Sales'])
ax2.set_xlabel('Radio',fontsize=12)
ax2.set_ylabel('Sales',fontsize=12)
ax3.scatter(x=df['Newspaper'],y=df['Sales'])
ax3.set_xlabel('Newspaper',fontsize=12)
ax3.set_ylabel('Sales',fontsize=12)
```

```python
correlation_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
x=df.drop('Sales', axis=1)
y=df['Sales']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
reg=LinearRegression()
reg.fit(x_train,y_train)
y_pred=reg.predict(x_test)
print(f'predict values;{y_pred}')
```

```python
from sklearn.metrics import r2_score,mean_absolute_error
mea=mean_absolute_error(y_test,y_pred)
print(f'Mean Absolute Error {mea:.2f}')
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)
print(f'r2 Score {r2:.2f}')
```

```python
plt.scatter(y_test, y_pred, label='Scatter Plot')
plt.xlabel('Actual Sales')
plt.ylabel('Predicted Sales')
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='--', linewidth=2, label='Linear Line')
plt.legend()
plt.title('Actual vs. Predicted Sales with Linear Line')
plt.style.use('seaborn-v0_8-darkgrid')
plt.show()

            
```
